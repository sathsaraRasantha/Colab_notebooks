```
import pandas as pd
data=pd.read_csv("clean_USA.csv")
data
```

```
brand_dummies=pd.get_dummies(data.brand)
model_dummies=pd.get_dummies(data.model)
title_dummies=pd.get_dummies(data.title_status)
color_dummies=pd.get_dummies(data.color)
state_dummies=pd.get_dummies(data.state)

data_new=data.drop(['Unnamed: 0'	,'price','brand','model','title_status','color','state','country'],axis='columns')
data_new
```

```
new_data=pd.concat([data_new,brand_dummies,model_dummies,title_dummies,color_dummies,state_dummies],axis='columns')
new_data
```

```
Y=data['price']
y=Y.to_numpy()
y
```

```
from sklearn.preprocessing import MinMaxScaler

scaler=MinMaxScaler()
new_data[['year','mileage']]=scaler.fit_transform(new_data[['year','mileage']])
new_data
```

```
import numpy as np

X=new_data.to_numpy()
X
```

```
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Activation,Dense
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.metrics import categorical_crossentropy
```

```
model=Sequential([
            Dense(units=16,activation='tanh'),
            Dense(units=32,activation='tanh'),
            Dense(units=1,activation='tanh')      

])
```

```
model.compile(optimizer=Adam(learning_rate=0.001),loss='categorical_crossentropy',metrics=['accuracy'])

model.fit(x=X,y=y,batch_size=10,epochs=30,shuffle=True)
```
