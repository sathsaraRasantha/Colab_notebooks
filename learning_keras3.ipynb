```
import pandas as pd
data = pd.read_csv('cardio_train (2).csv', delimiter=";", decimal=",")
#data=pd.DataFrame(data)
data
```

```
from sklearn.preprocessing import StandardScaler

data1=data.drop(['id','cardio'],axis='columns')
data1
```

```
from sklearn.model_selection import train_test_split

X=data1
Y=data['cardio']

X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
X_train.shape
```

```
from sklearn.preprocessing import MinMaxScaler

scaler=MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
print(X_train)
```

```
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Activation,Dense
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.optimizers import SGD
from tensorflow.keras.metrics import categorical_crossentropy
from keras.layers.advanced_activations import LeakyReLU
from tensorflow.keras.layers import Flatten, Dense
```

```
model=Sequential()
model.add(Dense(16))
model.add(LeakyReLU(alpha=0.05))
model.add(Dense(128, activation = 'softmax'))
model.add(Dense(128, activation = 'softmax'))
model.add(Dense(1, activation = 'sigmoid'))
                 
```

```
opt = SGD(lr=0.2, clipnorm=1.)
model.compile(optimizer=opt,loss='categorical_crossentropy',metrics=['accuracy'])

model.fit(x=X_train,y=Y_train,batch_size=10,epochs=100,validation_split=0.1,shuffle=True)
```

```
predictions=model.predict(X_test,batch_size=10)
predict=predictions.round()
print(predict)
print(Y_test)
```

```
from sklearn.metrics import confusion_matrix

cm=confusion_matrix(Y_test,predict)
print(cm)

#def plot_cof_matrix(cm,classes,normalize=Flase,title='Confusion Matrix',cmap=plt.cm.Blues):
 # plt.imshow(cm,interpolation='nearest',cmap=cmap)
```
